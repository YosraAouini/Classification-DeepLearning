# Generation of synthetic data

```python
#Import librairies
import pandas as pd
import random
import numpy as np
```

```python
#Set the seed the random number generator, which is useful for creating or random numbers that can be reproduced
random.seed(1254)
```

```python
#set the data size 
data_size=10000
#Create an empty datafame
data = pd.DataFrame()
```

```python
#Generate random numbers for students marks
np.random.seed(1544)
data['1st']  = np.random.uniform(2.89, 9, data_size).round(2)
data['2nd']  = np.random.uniform(3.25, 8.25, data_size).round(2)
data['3rd']  = np.random.uniform(3.48, 8.69, data_size).round(2)
data['4th']  = np.random.uniform(3.5, 9.03, data_size).round(2)
data['5th']  = np.random.uniform(4.01, 9.7, data_size).round(2)
```

```python
data['1st'].min(),data['2nd'].min(),data['3rd'].min(),data['4th'].min(),data['5th'].min()
```

```python
# Generate a list of random placement activities A,B,C,D,E
np.random.seed(14)
items = ['A', 'B', 'C', 'D', 'E']
proba_placement_activity=[0.1,0.17,0.32,0.27,0.14] # sum probabilities =1 
data['Placement_activity']=np.random.choice(items,data_size,p=proba_placement_activity)
   
```

```python
data.head(10)
```

```python
# Generate a list of F,M,Female and Male list 
np.random.seed(25)
Categories_Gender = ['M', 'Male', 'F', 'Female'] 
data['Gender']=random.choices(Categories_Gender, k=data_size)
```

```python
data.head(10)
```

```python
#Generate random College Code
np.random.seed(78)
colleges_codes = ['115', '116', '165', '241']
data['college_code']=random.choices(colleges_codes, k=data_size)
```

```python
data.head(10)
```

```python
#Generate a list of random Fields of Study
np.random.seed(98)
Fields_of_Study=['Engineering','Economics','Public Relations and Administration','Art','Environmental Science and Health','Aeronautics and Aviation Science']
proba=[0.25,0.19,0.11,0.1,0.24,0.11] # sum proba =1 
data['Fields_of_Study']=np.random.choice(Fields_of_Study,data_size,p=proba)
```

```python
data.head(10)
```

```python
#Generate random scholarship values
# Fisrt list = List of zeros, 
# Second list= List of values different of zero, between 122 and 15325 (These values are chosen))
np.random.seed(9875)
number_of_students_having_scholarship=3000
number_of_students_without_scholarship=data_size-number_of_students_having_scholarship # 10000-3000=7000 > list of zeros [0,0,0,]
scholarship1 = np.random.uniform(122, 15325, number_of_students_having_scholarship).round(2) # 3000
scholarship2= np.zeros(number_of_students_without_scholarship) #7000
scholarship=np.concatenate((scholarship1, scholarship2)) # 3000+7000 =10000
```

```python
scholarship[9100:9200]
```

```python
#Permutation of the created list
rng = np.random.default_rng(4)
data['scholarship']=rng.permutation(scholarship)
```

```python
#Export data
data.to_csv('data.csv',header=True,index=False)
```
